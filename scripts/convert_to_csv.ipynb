```python
import os

images = [os.path.join('data/final_data/', x) for x in os.listdir('data/final_data/') if x.endswith(".jpg")]
annots = [os.path.join('data/final_data/', x) for x in os.listdir('data/final_data/') if x.endswith(".txt")]

images.sort()
annots.sort()

len(images) == len(annots)
```

```python
csv_rows: list[list[str | int]] = []
```

```python
for annot, image in zip(annots, images):
    with open(annot, "r") as file:
        if len(file.buffer.read()) == 0:
            csv_rows.append([image] + [1,0,0])
        else:
            file.buffer.seek(0)
            class_for_img: list[int] = []
            negative = 0
            positive = 0
            for line in file:
                line = line.strip().split()
                if line[0] == '0':
                    negative += 1
                elif line[0] == '1':
                    positive += 1

            class_for_img = [0,negative,positive]
            csv_rows.append([image] + class_for_img)
```

```python
csv_rows[:5]
```

```python
import csv

CSV_FILE = "data/csv/road_classification.csv"
csv_rows = [["file_name", "normal", "negative", "positive"]] + csv_rows

with open(CSV_FILE, 'w') as f:
    writer = csv.writer(f)
    writer.writerows(csv_rows)
```
